Para extraer la info de las ciudades visitadas en USA existe una complicación extra: Existen ciudades con el mismo nombre en diferentes estados. Por ejemplo la Cuidad de Berkaley esta en 4 estados diferentes.

Para hacer el filtro usé la siguiente estrategia:
   - Paso 1 hice una lista con las cuidad-estado visitadas
   - Paso 2 hice un df solo con las cuidades de us. 
   - Paso 3 concatené la cuidad(city_ascii) y el estado (admin_name)
   - Paso 4 filtré de a cuerdo a los valores en la lista (1)
   - Paso 5 eliminar la columna creada para poder concatenar con los otros df


```python
# Paso 1: creación de listas con los nombres de las cuidades y lista de sus respectivos estados

list_visited_us = ['New York', 'Miami','Orlando', 'Atlanta', 'San Francisco','Berkeley', 'Boston' , 'Northampton', 'Baltimore',]
list_states             = ['New York','Florida','Florida', 'Giorgia','California','California','Massachusetts','Massachusetts','Maryland']
list_city_state_visited_us = ['New YorkNew York', 'MiamiFlorida','OrlandoFlorida', 'AtlantaGiorgia', 'San FranciscoCalifornia','BerkeleyCalifornia', 'BostonMassachusetts' , 'NorthamptonMassachusetts', 'BaltimoreMaryland']
```

```python
#Paso 2: df solo con las cuidades de us

df_us = df[df['country']=='United States']
```

```python
# Paso 3: concatenación de la cuidad(city_ascii) y el estado (admin_name)

df_us['city_state']=df_us['city_ascii'] + df_us['admin_name']
```

```python
# Paso 4: filtrar los valores en la nueva columna de acuerdo a los de la lista creada 

df_us =df_us.loc[df_us['city_state'].isin(list_city_state_visited_us)]
```

```python
#Paso 5: drop la nueva columna
df_visited_us = df_us.drop('city_state',axis=1)
df_visited_us 
```

## Concatenación de los set de datos creados

```python
#convine all the data frames
#result = pd.concat(frames)
list_df=[df_visited_cl , df_visited_per, df_visited_mx, df_visited_bg,df_visited_tr, df_visited_it ,df_visited_es ,df_visited_de , df_visited_us,]
df_visited_cities = pd.concat(list_df)
df_visited_cities
```

### Mapa

Para crear un mapan usando folium debemos instanciar un objeto Map

```python
#create un mapa
my_map = folium.Map(zoom_start=2,
    location=[13.133932434766733, 16.103938729508073])
```

```python
my_map
```

```python
# my_map is a map object that is empty and now I will add a city 
```

```python
# create a city
city = df_visited_cities.loc['Lima']  #Crea una serie con la info de la cuidad

# create a marker on the map with the city indicated

folium.Marker(
    location=[city["lat"],city["lng"]]).add_to(my_map)
my_map
```

```python
#añadir todas las ciudades
```

```python
my_map = folium.Map(
    zoom_start=2,
    location=[13.133932434766733, 16.103938729508073])

for _, city in df_visited_cities.iterrows():
    folium.Marker(
    location=[city["lat"],city["lng"]], # aparecen las ciudades del data set
    popup=city['city_ascii'],                 # the name of the city appear when you click on it
    tooltip=city['city_ascii']                # the name of the city pops when you are on it
    ).add_to(my_map)
    
    
```

## Personalizar el mapa. 
Voy a añadir color a los pins:
- Morado para las cuidades en las que he vivido.
- Azul para las que he visitado 


```python
#lista de ciudades en las que he vivido
list_lived=['Santiago','Rancagua','Northampton','Sofia','Miami']

#crear columan con la condición

df_visited_cities['lived']=np.where(df_visited_cities['city_ascii'].isin(list_lived), 1,0)

df_visited_cities['color']=np.where(df_visited_cities['lived']==1, 'purple','blue')
```

```python
my_map = folium.Map(
    zoom_start=2,
    location=[13.133932434766733, 16.103938729508073])

for _, city in df_visited_cities.iterrows():
    folium.Marker(
    location=[city["lat"],city["lng"]], # aparecen las ciudades del data set
    popup=city['city_ascii'],                 # the name of the city appear when you click on it
    tooltip=city['city_ascii'],                # the name of the city pops when you are on it
    icon=folium.Icon(color=city['color'])
    ).add_to(my_map)
   
```
